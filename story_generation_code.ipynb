```python
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer, GenerationConfig
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
def image_to_text(url):
  pipe = pipeline("image-to-text", model="Salesforce/blip-image-captioning-base")
  generated_text = pipe(url)[0]["generated_text"]
  return generated_text
```

```python
tokenizer = AutoTokenizer.from_pretrained("bigscience/bloomz-1b7")
model = AutoModelForCausalLM.from_pretrained("bigscience/bloomz-1b7").to(device)
```

```python
pipe = pipeline("image-to-text", model="Salesforce/blip-image-captioning-base")
```

```python
def generate_story_from_image(url):
    pipe = pipeline("image-to-text", model="Salesforce/blip-image-captioning-base")
    generated_text = pipe(url)[0]["generated_text"]

    # Use the generation config to generate text from the model
    inputs = tokenizer.encode(f'Once upon a time, {generated_text}', return_tensors="pt").to(device)
    output = model.generate(inputs, min_length=100,
    max_length=150,
    top_p=0.8,
    top_k=40,
    temperature=0.8,
    no_repeat_ngram_size=2,
    num_beams=4,
    early_stopping=True,
    do_sample=True,
    use_cache=False,)

    # Decode the output and return it
    generated_story = tokenizer.decode(output[0], skip_special_tokens=True)
    return generated_story
```

```python
generated_story = generate_story_from_image("/content/sample1.jpeg")
print(generated_story)
```
